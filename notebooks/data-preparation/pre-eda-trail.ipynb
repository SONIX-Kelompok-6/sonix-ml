# Clearing Brand

```python
df["Brand"] = (
    df["Brand"]
    .astype(str)
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
    .str.title()
)

brand_uniques = df["Brand"].dropna().unique()
# brand_uniques.sort()
brand_uniques = sorted(brand_uniques)
print(f"Brand ({len(brand_uniques)} uniques): \n",brand_uniques)

df["Brand"].value_counts()
```

# Cleaning Terrain

```python
df["Trail terrain"] = (
    df["Trail terrain"]
    .astype(str)
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
    .str.title()
)

terrain_uniques = df["Trail terrain"].dropna().unique()
# terrain_uniques.sort()
terrain_uniques = sorted(terrain_uniques)
print(f"Trail terrain ({len(terrain_uniques)} uniques): \n",terrain_uniques)

df["Trail terrain"].value_counts().head(20)
```

```python
terrain_map = {
    "Light": "Light",
    "Moderate": "Moderate",
    "Technical": "Technical",
    "Lightmoderate": "Light|Moderate",
    "Moderatetechnical": "Moderate|Technical",
}

df["terrain_norm"] = df["Trail terrain"].map(terrain_map)

df.head()
```

```python
unmapped = df[df["terrain_norm"].isna()]["Trail terrain"].value_counts()
print("UNMAPPED terrain values:\n", unmapped)
# assert df["terrain_norm"].notna().all()
```

```python
terrain_norms_unique = df["terrain_norm"].dropna().unique()
print(f"terrain - norm : ({len(terrain_norms_unique)}) uniques\n", terrain_norms_unique)


df["terrain_norm"].value_counts()
```

```python
df["terrain_lists"] = df["terrain_norm"].str.split("|")
df.head()
```

```python
terrain_exploded = df["terrain_lists"].explode()

terrain_ohe = (
    pd.crosstab(terrain_exploded.index, terrain_exploded)
    .reindex(df.index, fill_value=0)   
)

terrain_ohe = terrain_ohe.rename(columns={
    "Light": "terrain_light",
    "Moderate": "terrain_moderate",
    "Technical": "terrain_technical"
})

# gabung ke df
df = pd.concat([df, terrain_ohe], axis=1)
```

```python
df.head()
```

```python
# make sure value-nya bener 0/1
for c in ["terrain_light", "terrain_moderate", "terrain_technical"]:
    assert set(df[c].unique()).issubset({0, 1})
```

```python
print(df[df["Trail terrain"]=="Light"][["Trail terrain","terrain_norm","terrain_light","terrain_moderate","terrain_technical"]].head())
print(df[df["Trail terrain"]=="Moderate"][["Trail terrain","terrain_norm","terrain_moderate","terrain_technical","terrain_light"]].head())
```

```python
df["Season"] = (
    df["Season"]
    .astype(str)
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
    .str.title()
)

heel_counter_stiffness_uniques = df["Season"].dropna().unique()
# heel_counter_stiffness_uniques.sort()
heel_counter_stiffness_uniques = sorted(heel_counter_stiffness_uniques)
print(f"Season ({len(heel_counter_stiffness_uniques)} uniques): \n",heel_counter_stiffness_uniques)

df["Season"].value_counts()
```

```python
season_map = {
    "All Seasons": "All",
    "Summerall Seasons": "Summer|All",
    "Winter": "Winter",
    "-": pd.NA,
}

df["season_norm"] = df["Season"].map(season_map)
```

```python
unmapped = df[df["season_norm"].isna() & df["Season"].ne("-")]["Season"].value_counts()
print("UNMAPPED Season values (should be empty):\n", unmapped)
# assert unmapped.empty
```

```python
season_norm_unique = df["season_norm"].unique()
print(f'Season - norm: ({len(season_norm_unique)}) uniques\n', season_norm_unique)
print(df["season_norm"].value_counts())
```

```python
df["season_list"] = df["season_norm"].str.split("|")

season_exploded = df["season_list"].explode()

season_ohe = (
    pd.crosstab(season_exploded.index, season_exploded)
    .reindex(df.index, fill_value=0)
)

season_ohe = season_ohe.rename(columns={
    "All": "season_all",
    "Summer": "season_summer",
    "Winter": "season_winter"
})

for col in ["season_all","season_summer","season_winter"]:
    if col not in season_ohe.columns:
        season_ohe[col] = 0

season_ohe = season_ohe[["season_all","season_summer","season_winter"]]

df = pd.concat([df, season_ohe], axis=1)
df.head()
```

```python
df = df.loc[:, ~df.columns.duplicated()]

temp_water = (
    df["Waterproofing"]
    .astype(str)
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)

water_ohe = pd.get_dummies(temp_water, prefix="water").astype(int)

mapping = {
    "water_Waterproof": "water_proof",
    "water_Water repellent": "water_repellent",
    "water_WaterproofWater repellent": "water_both", # atau bisa dipisah nanti
    "water_-": "water_none"
}

water_ohe = water_ohe.rename(columns=mapping)

target_cols = ["water_proof", "water_repellent", "water_both", "water_none"]

for col in target_cols:
    if col not in water_ohe.columns:
        water_ohe[col] = 0


df = df.drop(columns=[c for c in target_cols if c in df.columns])

df = pd.concat([df, water_ohe[target_cols]], axis=1)

print("Pengecekan Hasil Waterproofing:")
sample_check = df[df["Waterproofing"].isna() | (df["Waterproofing"] == "-")].head(5)
print(sample_check[["Waterproofing"] + target_cols])
```
