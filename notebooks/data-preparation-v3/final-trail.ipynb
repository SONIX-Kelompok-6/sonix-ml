## Separate Brand-Name

```python
''' 
Running shoes for trail brand in our dataset include:
    Adidas
    Altra
    ASICS
    Brooks
    HOKA
    Icebug
    Inov8
    Kailas
    KEEN
    La Sportiva
    Merrell
    New Balance
    Nike
    NNormal
    On
    Salomon
    Saucony
    Topo
    Xero
    Scarpa
    The North Face
'''

brands_list = [
    "Adidas", "Altra", "ASICS", "Brooks", "HOKA", "Icebug", "Inov8", 
    "Kailas", "KEEN", "La Sportiva", "Merrell", "New Balance", "Nike", 
    "NNormal", "On", "Salomon", "Saucony", "Topo", "Xero", "Scarpa", 
    "The North Face"
]
brands = [b.lower() for b in brands_list]
brands.sort(key=len, reverse=True)
```

```python
def split_brand_name(full_text):
    for brand in brands:
        if full_text.startswith(brand):
            # Sisa dari brand dijadiin name semua
            name = full_text[len(brand):].strip()
            return brand, name
    return "Unknown", full_text 


df_ori[['brand', 'name']] = df_ori['brand-name'].apply(lambda x: pd.Series(split_brand_name(x)))

# Atur urutan kolom agar brand dan name tetap ada di depan
cols = ['brand', 'name'] + [c for c in df_ori.columns if c not in ['brand', 'name', 'brand-name']]
df_ori = df_ori[cols]

df_ori.head()
```

```python
df['waterproofing'] = df['waterproofing'].astype(str).str.lower()
base_water = ['waterproof', 'water repellent']

def check_not_waterproof(val):
    if val in ['-', '0', 'nan', 'none']:
        return 1
    return 0

for level in base_water:
    column_name = level.replace(' ', '_')
    
    if level == 'not waterproof':
        df[column_name] = df['waterproofing'].apply(check_not_waterproof)
    else:
        df[column_name] = df['waterproofing'].str.contains(level, na=False).astype(int)
        df.loc[df['waterproofing'].isin(['-', '0']), column_name] = 0

print("Rows:", len(df))

water_cols = [l.replace(' ', '_') for l in base_water]

print("\nSample Comparison:")
print(df[["waterproofing"] + water_cols].head(10))
```
